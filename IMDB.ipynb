```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results_with_crew.csv')
```

```python
#### Most Frequent Directors ####
```

```python
df['directors'] = df['directors'].fillna('Unknown')
top_directors = df['directors'].value_counts().head(10)
sns.barplot(x=top_directors.values, y=top_directors.index, hue=top_directors.index, palette='viridis',legend=False)
plt.title('Top 10 Directors by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
plt.show()
```

```python
#### Average Movie Ratings per Genre ####
```

```python
df['genres'] = df['genres'].fillna('Unknown')
avg_rating = df.groupby('genres')['averageRating'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=avg_rating.values, y=avg_rating.index,hue=avg_rating.index, palette='magma')
plt.title('Top Genres by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.show()
```
